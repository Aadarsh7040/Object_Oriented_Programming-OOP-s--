# Generator

# Difference between yield keyword and Return statement

```python
# yield keyword
1. Code will not stop and continue with next iteration.
2. yield returns the generator object.
3. Run multiple times.
4. Returns multiple values from a function

# Return keyword
1. Code will stop(not execute) after return statement.
2. return statement returns the result to the caller.
3. Runs only a single time.
4. Returns only a single value from a function
```

```python
# 1.range 1 to 10

def Testgenerator():
    for  i in range(1,10):
        return i
    
Testgenerator()
```

```python
def Testgenerator():
    for  i in range(1,10):
        yield i
    
for i in Testgenerator():
    print(i)
```

```python
# 2.squares of item in a list

def testgenerator(list1):
    for i in list1:
        return i**2
    
testgenerator([1,2,3,4,5])
```

```python
def testgenerator(list1):
    for i in list1:
        yield i**2
    
for j in testgenerator([1,2,3,4,5]):
    print(j)
```

```python
# String

def testgenerator(string1):
    for i in string1:
        return i
    
testgenerator('python') 
```

```python
def testgenerator(string1):
    for i in string1:
        yield i
    
for j in testgenerator('python'):
    print(j)
```

```python
# range with string

def testgenerator(string1):
    for i in range(len(string1)):
        print(f'Testgenerator {i+1}')
        return string1[i]
    
testgenerator('python')
```

```python
def testgenerator(string1):
    for i in range(len(string1)):
        print(f'Testgenerator {i+1}')
        yield string1[i]
    
for j in testgenerator('python'):
    print(j)
```

```python
# square and cube of a given number

def testgenerator(num):
    square = num**2
    return square

    cube = num**3
    return cube

testgenerator(4)
```

```python
def testgenerator(num):
    square = num**2
    yield square

    cube = num**3
    yield cube

for j in testgenerator(4):
    print(j)
```

```python
# Fibonacci

def fibonacci(num):
    first, second = 0,1
    for i in range(num):
        first, second = second, first + second
        return first
        
fibonacci(5)
```

```python
def fibonacci(num):
    first, second = 0,1
    for i in range(num):
        first, second = second, first + second
        yield first
        
for i in fibonacci(5):
    print(i)
```
